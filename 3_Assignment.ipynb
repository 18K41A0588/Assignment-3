5Q) write a NumPy program to compute the sum of the diagonal element of a given array(matrix).

```
import numpy as np
```

```
a=[[5,2,3],[1,8,2],[9,2,5]]
b=[[6,3,4],[2,2,1],[8,4,1]]
c=np.diagonal(a)
d=np.diagonal(b)
print("Array 1 :")
for i in range(0,len(a)):
    print(a[i])
print("Array 2 : ")
for i in range(0,len(b)):
    print(b[i])
    
print("The diagonal Elements of first array:",c)
print("The diagonal Elements of second array:",d)
e=c+d
print("The sum of diagonal elements is : ",e)
```

4Q)  Write a NumPy program to compute the condition number of a given matrix.

```
import numpy.linalg as nl
m=np.array([[1,2],[3,4]])
print("Input matrix is : ")
print(m)
print("Condition number of the said matrix :")
print(nl.cond(m))
```
